```python
import pandas as pd
mdf = pd.read_csv('data/linked_mortality_file_1999_2000.csv') # Load mortality file
print( mdf.head() )
gdf = pd.read_sas("data/DEMO.xpt", format="xport") # Load demographics file
print( gdf.head() )
df = gdf.merge(mdf, on="SEQN", how="inner") # Merge mortality and demographics on SEQN variable
```

## Variables Described

**ELIGSTAT**: Tells us whether someone can be tracked in the mortality data. Only respondents marked as eligible should be included when studying death outcomes.

**MORTSTAT**: Tells us whether a person is alive or dead by the end of the follow-up period. Answers whether the person is dead or not. 

**PERMTH_INT**: Measures how much time passed (in months) between the person’s NHANES interview and the outcome we observe. If the person died, this is the time until death. If the person is still alive, it measures how long they were followed before the study ended.

**RIDAGEEX**: Tells us the person’s age in months at the time of the NHANES exam.
